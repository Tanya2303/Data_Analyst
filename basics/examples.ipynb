## Manage a todo list

```python
todo_list = ["wake_up", "exercise", "study", "code"]

##adding a new task
todo_list.append("read")
todo_list.append("practice")

##removing a completed task
todo_list.remove("exercise")

##checking if a task is in the list
if "study" in todo_list:
    print("Task 'study' is in the todo list.")

print('task in todo list:')
for task in todo_list:
    print('-', task)
# A simple Python program to manage a todo list
```

## Organising student grades

```python
grades = [85, 92, 78, 90, 88]

grades.append(95)  # Adding a new grade

average_grade = sum(grades) / len(grades)
print(f"Average grade: {average_grade}")

high_achievers = [grade for grade in grades if grade >= 90]
print("High achievers (90 and above):", high_achievers)

low_achievers = [grade for grade in grades if grade < 80]
print("Low achievers (below 80):", low_achievers)

# A simple Python program to organise student grades
```

## Managing an inventory system

```python
inventory = ["apple", "banana", "orange"]
inventory.append("grape")  # Adding a new item
inventory.remove("banana")  # Removing an item

item_to_check = "orange"
if item_to_check in inventory:
    print(f"{item_to_check} is in stock.")
else:
    print(f"{item_to_check} is out of stock.")

print("Current inventory:")
for item in inventory:
    print("-", item)
# A simple Python program to manage an inventory system
```
